````python
# Imports y configuración
import sys
from pathlib import Path

# Permite importar config y módulos del proyecto desde notebooks/
sys.path.append(str(Path.cwd().parent))

import pandas as pd
import matplotlib.pyplot as plt

from config import CSV_PROCESADO, CIUDADES, get_categoria_aqi

pd.set_option("display.max_columns", None)
print("Ciudades monitoreadas:", list(CIUDADES.keys()))
````

## 1. Carga del dataset procesado

````python
df = pd.read_csv(CSV_PROCESADO)
df["timestamp"] = pd.to_datetime(df["timestamp"], utc=True, errors="coerce")
print("Filas:", len(df))
df.head()
````

## 2. Información general y valores nulos

````python
df.info()
print("\nValores nulos por columna:")
print(df.isna().sum())
````

## 3. Estadísticas descriptivas

````python
df.describe(numeric_only=True)
````

## 4. AQI promedio por ciudad

````python
aqi_ciudad = df.groupby("ciudad")["aqi"].mean().sort_values(ascending=False)
print(aqi_ciudad)

ax = aqi_ciudad.plot(kind="bar", color="#00964F", figsize=(8, 4))
ax.set_title("AQI promedio por ciudad")
ax.set_ylabel("AQI")
plt.tight_layout()
plt.show()
````

## 5. Distribución de categorías AQI

````python
if "categoria_aqi" in df.columns:
    conteo = df["categoria_aqi"].value_counts()
    print(conteo)
    conteo.plot(kind="pie", autopct="%1.1f%%", figsize=(5, 5))
    plt.ylabel("")
    plt.title("Distribución de categorías AQI")
    plt.show()
````

## 6. Correlación entre variables numéricas

````python
numericas = df.select_dtypes(include="number")
corr = numericas.corr()

fig, ax = plt.subplots(figsize=(9, 7))
cax = ax.imshow(corr, cmap="RdBu_r", vmin=-1, vmax=1)
ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=90)
ax.set_yticklabels(corr.columns)
fig.colorbar(cax)
plt.title("Matriz de correlación")
plt.tight_layout()
plt.show()
````

## 7. Evolución temporal de PM2.5

````python
fig, ax = plt.subplots(figsize=(10, 5))
for ciudad, grupo in df.sort_values("timestamp").groupby("ciudad"):
    ax.plot(grupo["timestamp"], grupo["pm25"], marker="o", label=ciudad)
ax.set_title("Evolución de PM2.5 por ciudad")
ax.set_xlabel("Tiempo")
ax.set_ylabel("PM2.5")
ax.legend()
plt.tight_layout()
plt.show()
````
